```python
from pylab import *  
mpl.rcParams['font.sans-serif'] = ['SimHei'] #指定默认字体  
mpl.rcParams['axes.unicode_minus'] = False #解决保存图像是负号'-'显示为方块的问题 
import pandas
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
zhfont = mpl.font_manager.FontProperties(fname='/usr/share/fonts/truetype/droid/DroidSansFallbackFull.ttf')


def barplot(sorted_freq, title):
    print(sorted_freq)
    sorted_freq = sorted_freq[:10]
    figure, ax = plt.subplots(figsize=(15, 10))
    ind = np.arange(0,1*len(sorted_freq),1)
    rects1 = ax.bar(ind, [i[1] for i in sorted_freq], 0.5, color='r')
    ax.set_title(title, fontproperties=zhfont, fontsize=15)
    ax.set_xticks(ind + 0.5/2)
    ax.set_xticklabels([i[0] for i in sorted_freq], fontproperties=zhfont, fontsize=15)
    #print([i[0] for i in sorted_freq])

    return

df = pandas.read_csv('070601.csv')

def barplotbyidx(idx):
    #print(k)
    k = df.keys()[1:][idx]
    names = df[k]
    freq = Counter(names)
    sorted_freq = sorted(freq.items(), key=lambda i: i[1], reverse=True)
    barplot(sorted_freq, k)
    #print(sorted_freq)
```
